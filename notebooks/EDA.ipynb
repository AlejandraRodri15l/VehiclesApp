```python
car_data = pd.read_csv('../vehicles_us.csv') # leer los datos
fig = px.scatter(car_data, x="odometer", y="price") # crear un gráfico de dispersión
fig.show() # crear gráfico de dispersión
```

```python
print(car_data.head(30)) # Muestra los primeros 30 lineas del DataFrame
```

```python
print(car_data.info(show_counts=True)) # mostrar información del DataFrame
print()
print(car_data.isna().sum())# Identificar los valores ausentes de los DataFrame
print()
print(car_data.duplicated().sum()) # Para conocer la cantidad de filas duplicadas.
```

```python
# Completa los nombres de productos ausentes de is_4wd y paint_color
car_data['is_4wd'] = car_data['is_4wd'].fillna(value= 0)
car_data['paint_color'] = car_data['paint_color'].fillna(value='Unknown')
print(car_data.head())
```

```python
# Calcular el promedio de model_year para cada model
model_year_means = car_data.groupby('model')['model_year'].median()

# Definir una función para reemplazar los valores ausentes con el promedio correspondiente
def fill_missing_model_year(row):
    if pd.isna(row['model_year']):
        return model_year_means[row['model']]
    else:
        return row['model_year']

# Aplicar la función a cada fila del DataFrame
car_data['model_year'] = car_data.apply(fill_missing_model_year, axis=1)

# Verificar los resultados
print(car_data.head())
```

```python
# Calcular el promedio de cylinders  para cada model
model_year_means = car_data.groupby('model')['cylinders'].mean()

# Definir una función para reemplazar los valores ausentes con el promedio correspondiente
def fill_missing_cylinders(row):
    if pd.isna(row['cylinders']):
        return model_year_means[row['model']]
    else:
        return row['cylinders']

# Aplicar la función a cada fila del DataFrame
car_data['cylinders'] = car_data.apply(fill_missing_cylinders, axis=1)

# Verificar los resultados
print(car_data.head())
```

```python

# Calcular el promedio de 'odometer' para cada 'model' y redondear a dos decimales
model_odometer_means = car_data.groupby('model')['odometer'].mean().round(2)

# Función para rellenar valores ausentes en 'odometer' con el promedio del modelo
def fill_missing_odometer(row):
    if pd.isna(row['odometer']):
        return model_odometer_means[row['model']]
    else:
        return row['odometer']

# Aplicar la función a cada fila del DataFrame
car_data['odometer'] = car_data.apply(fill_missing_odometer, axis=1)

car_data['odometer'] = car_data['odometer'].fillna(value= 500000) #Remmplazar con un valor fijo ya que un modelo completo no se diligencio el cuentakilómetros
print(car_data[car_data['odometer'].isna()])
```

```python
# Crear una nueva columna que nos permita analizar si el vehiculo esta en optimas condiciones de acuerdo al kilometraje y su antiguadad

# Convertir 'date_posted' a formato de fecha
car_data['date_posted'] = pd.to_datetime(car_data['date_posted'], errors='coerce')

# Extraer el año de 'date_posted'
car_data['year_posted'] = car_data['date_posted'].dt.year

# Crear una nueva columna que calcule la diferencia en años entre 'date_posted' y 'model_year'
car_data['age_at_posting'] = car_data['year_posted'] - car_data['model_year']

# Convertir a entero varias columnas para posteriores calculos
car_data['age_at_posting'] = car_data['age_at_posting'].astype(int)
car_data['cylinders'] = car_data['cylinders'].astype(int)
car_data['model_year'] = car_data['model_year'].astype(int)
car_data['odometer'] = car_data['odometer'].astype(int)


print(car_data.head(30))
```

```python
# Extraer la marca de carro de la columna 'model'
car_data['brand'] = car_data['model'].str.split().str[0]

# Mostrar el DataFrame con la marca de carro extraída
print(car_data)
```

```python
print(car_data.info())
```

```python
analysis_condition = car_data.groupby(['model_year','condition']).agg({'price':'mean', 'odometer':'mean'}).reset_index()

print(analysis_condition.head(30))
```

```python
# Crear el gráfico de barras agrupado
fig = px.scatter(analysis_condition, x='model_year', y='price', size='odometer',
                color='condition', title='Precio de vehiculo de acuerdo al año y kilometraje de acuerdo a su condición')

# Mostrar el gráfico
fig.show()
```

```python
# Calcular el precio promedio de acuerdo a odometer y brand
avg_price_by_odometer_brand = car_data.groupby(['brand', 'odometer']).agg({'price': 'mean'}).reset_index()

# Crear el histograma
fig = px.histogram(avg_price_by_odometer_brand, x='odometer', y='price', color='brand', 
                barmode='group', title='Precio Promedio según Kilometraje y Marca',
                labels={'odometer': 'Kilometraje', 'price': 'Precio Promedio', 'brand': 'Marca'})

# Mostrar el gráfico
fig.show()
```
